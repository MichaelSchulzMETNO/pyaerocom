### Installation of pyaerocom (detailed instructions)

In the following, detailed installation instructions are provided, that should help you getting pyaerocom running on your machine, including all dependencies.

**Note**: 
- Pyaerocom is programmed and tested in Python version 3.6 (or later). If you use an older version of Python (e.g. Python 2.7) we recommend installing Python 3 and run  in a second environment (e.g. using Anaconda, see next point).
- The following instructions assume that you use Python 3 and [Anaconda](https://www.anaconda.com/) as a package manager. 

#### Short version

Make sure you have [all dependencies](https://github.com/metno/pyaerocom#requirements) installed, download [latest release](https://github.com/metno/pyaerocom/releases) (or download and unzip, or clone [repo](https://github.com/metno/pyaerocom)), navigate into the source tree folder (named *pyaerocom*, contains *setup.py* file) and call:

`python setup.py install`

#### Installation of Python using Anaconda

We recommend using Anaconda as a package manager as it provides a powerful interface for installing Python packages and keep them up to date. It further allows to create multiple environments (e.g. Python 2 and Python 3).

If you already have Anaconda (with Python >= 3.6) installed on your machine, you can skip this point. 

##### I have an Anaconda installation but with Python 2

If you have already have Anaconda installed but with Python 2 we recommend you to create a new Python 3 environment using 

`conda create -n py3 python=3`

in your command line (should work in all operating systems). Note, you may also choose another name than `py3` for your environment. Activate your new environment from the command line using 

 - **Windows**  
 
 `activate py3`

- **Linux, macOS**  

  `source activate py3`

Then continue with installing all requirements as introduced in the following sections. 

For details related to conda enviroments, [see here](https://conda.io/docs/user-guide/tasks/manage-environments.html#).


##### Installing Anaconda with Python 3 from scratch

Please follow the [instructions provided on the Anaconda website](https://www.anaconda.com/download/) (for your OS, respecively).

You may go for Miniconda or the full installation (Anaconda) and continue with installing the missing requirements.

#### Installing dependencies

You can find a list of dependencies for pyaerocom [here](https://github.com/metno/pyaerocom#requirements).

##### Checking for  installed pacakages

To list all installed packages use

`conda list`

To check if a specific package is installed, use

`conda list <package_name>`

##### Installing packages with conda

Most Python packages are 'known' by Anaconda and can be installed using

`conda install <package_name>`

If this does not work, it usually helps to google for "conda install <package_name>" and you will find a solution (usually the package will be available in another channel, such as conda-forge).

#### Installing pyaerocom requirements

##### Using provided *environment.yml* file

Note: this is the easiest way to get all requirements. It [creates a new conda environment](https://conda.io/docs/user-guide/tasks/manage-environments.html#creating-an-environment-from-an-environment-yml-file) with name *pya* by calling 

```
conda env create -f environment.yml
```

Activate this environment using

- **Windows**  
 
 `activate pya`

- **Linux, macOS**  

  `source activate pya`
  
 and continue with installation of pyaerocom as indicated below.


##### Installing libraries on my own

If you do not have it already installed, start with installing the *iris* package using:

`conda install -c conda-forge iris`

This will install or update most of the pyaerocom dependencies such as numpy, pandas, matplotlib, NetCDF4, cf_units.

Continue with the installation of the other dependencies (that did not come together with iris):

`conda install xarray`  
`conda install cartopy`  
`conda install matplotlib=2.2.2`  
   
The last command downgrades matplotlib to version 2.2.2. due to [this known issue](https://github.com/SciTools/cartopy/issues/1120) of matplotlib 3.0.0 (version when this was written) with cartopy. 

And the optional dependencies:

`conda install seaborn`

`conda install geopy`

#### Install pyaerocom

Download and unzip [latest release](https://github.com/metno/pyaerocom/releases) (or download and unzip, or clone [repo](https://github.com/metno/pyaerocom)), navigate into the source tree folder (named *pyaerocom*, contains *setup.py* file) and call:

`python setup.py install`

If you want to install in [development mode](https://setuptools.readthedocs.io/en/latest/setuptools.html#development-mode), call 

`python setup.py develop` 

rather than `python setup.py install`.
